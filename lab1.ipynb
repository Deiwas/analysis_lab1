# Обнаружение дубликатов
Проверим датасет на наличие дубликатов - полностью повторяющихся строчек

```python
print(df.duplicated().sum())
```

Дубликаты были обнаружены, поэтому необходимо от них избавиться с помощью метода drop_duplicates(). После также была обновлена индексация датафрейма.

```python
df = df.drop_duplicates()
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates().reset_index()
df.head(20)
```

Неявные дубликаты можно найти с помощью метода unique().

```python
print(df['Name'].unique())
```

# Поиск пропусков в таблице
Далее будет проведен поиск пропусков в таблице.

```python
print(df.isna().sum())
```

Поскольку SellingPrice является важным столбцом датафрейма, все строки без стоимости машины придется удалить. Это решает и остальные проблемы с пустыми ячейками. Поскольку количество пропусков значительно ниже общего количества строк, это никак не повредит данным выборки.

```python
df = df.dropna(subset=['SellingPrice'])
print(df.isna().sum())
```

# Работа с типами данных
Проведена проверка на соответствие всех данных действительности. Как видно, столбцу year соответствует текстовый тип.

```python
df.dtypes
```

При переводе столбца даты в соответствующий тип возникла проблема: кавычки не позволяли осуществить конверсию. Поэтому был создан метод, удаляющий все кавычки.

```python
df['year'] = df['year'].str.replace("'","")
```

```python
df['year'] = pd.to_numeric(df['year'])
```

```python
df
```

# Созднание сводных таблиц
Далее будет построена сводная таблица по данным этого датафрейма.

```python
dft = df.pivot_table(index='year', columns='owner',values='SellingPrice', aggfunc='sum')
dft = dft.fillna(0)
dft
```

В этой таблице показаны суммы продаж на тип владельцев на протяжении всех лет продаж. Исходя из данных, количество первых владельцев авто снижается к 2020 году, набрав свой пик в 2017 году. То же относится и к сумме продаж вторых владельцев авто, где после 2012-2014 года наблюдается резкий спад. При этом, в начиная с 2017 года набирают популярность машины с тест-драйва.

Проведу еще две группировки.

```python
dft = df.pivot_table(index='fuel', columns='transmission',values='kmdriven', aggfunc='sum')
dft = dft.fillna(0)
dft
```

Исходя из этой таблицы можно сделать выводы о значительном преобладании ручной трансмисии над автоматической. Однако, автоматическая трансмиссия незначительно отстает от ручной в дизельном топливе и бензине, а также преобладает в электрическом топливе.

```python
dft = df.pivot_table(index='year', columns='fuel', values='kmdriven', aggfunc=np.mean)
dft = dft.fillna(0)
dft
```

# Дополнительное задание

Выделить пять лет с максимальными суммарными продажами. Для этих годов создать сводную таблицу со средней ценой продажи по типам fuel.

```python
df2 = df.groupby('year')['SellingPrice'].sum().sort_values(ascending = False).head(5)
best_years = df2.index.values.tolist()
df2.to_frame()
```

Создание сводной таблицы по типам fuel для 5 лет с максимальными суммами

```python
pd.pivot_table(df[df.year.isin(best_years)], index=['year', 'fuel'], values='SellingPrice', aggfunc=np.mean)
```

```python
pd.pivot_table(df[df.year.isin(best_years)], index=['year', 'fuel'], values='SellingPrice', aggfunc='sum')
```

```python
pd.pivot_table(df[df.year.isin(best_years)], index=['year', 'fuel'], values='SellingPrice', aggfunc='count')
```

В этой таблице демонстрируется количество машин на каждом топливе по годам выборки. Видно значительное преобладание машин на дизельном топливе и бензине, с наибольшим их количеством, приходящимся на 2011-2018 годы.
